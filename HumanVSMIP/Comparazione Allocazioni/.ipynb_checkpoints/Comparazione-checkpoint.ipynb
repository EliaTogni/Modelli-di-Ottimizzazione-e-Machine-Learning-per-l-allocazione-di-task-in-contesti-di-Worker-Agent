```python
import pandas as pd
import numpy as np

#Metodo che taglia la parte errata da ogni stringa
def sub(DataFrame):
    lista = [i[0:36] for i in DataFrame]
    return lista

df = pd.read_csv("../../Dataset/Decisions.csv")

#Tolgo la parte errata dalle stringhe della seconda colonna
dg = df.iloc[:, 1]
lista1 = sub(dg)
    
#Appendo le nuove colonne al DataFrame
lista1 = pd.DataFrame(lista1)
lista1.columns = ["temp1"]
df = df.join(lista1)
    
#Rimuovo la vecchia colonna e rinomino quella nuova
df = df.drop("Session ID", axis = 1)
df = df.rename(columns = { "temp1" : "Session ID"})

df = df.iloc[:, [7, 1, 2, 5]]
df
```

Le sessioni di gioco prese in considerazione sono scelte tra le game session con score più alto tra quelle che self reportano la strategia analizzata.<br />
Per ciascuna analisi, si è scelto di confrontare l'allocazione ritenuta ideale per quella strategia, l'allocazione effettuata dal modello e l'allocazione delle task del giocatore.<br />
La bontà di ogni allocazione è valutata tramite score percentuale, calcolato come value delle task portate a termine con successo diviso value totale delle task.<br />
